```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# напишіть функцію гіпотези лінійної регресії у векторному вигляді;
"""
В лінійній регресії гіпотеза може бути представлена у векторному вигляді як:
ho(x) = θt * x
де:
ho(x) - прогнозоване значення;
θ - вектор параметрів моделі (також відомий як вектор ваг);
х - вектор ознак.
Ось функція для реалізації цієї гіпотези у векторному вигляді в Python:
Ця функція приймає вектор параметрів О і вектор ознак
x і обчислює прогнозоване значення, використовуючи їх дотичний добуток.
"""
def linear_hypothesis(w, X):
    """
    Parameters:
    w (numpy.ndarray): Vector of parameters.
    X (numpy.ndarray): Matrix of features.

    Returns:
    numpy.ndarray: Predicted values.
    """
    return np.dot(X, w)
```

```python
# створіть функцію для обчислення функції втрат у векторному вигляді;
"""
Для обчислення функції втрат у векторному вигляді
в лінійній регресії ми можемо використовувати середньоквадратичну помилку
(MSE - Mean Squared Error), яка визначається наступним чином:
J(θ) = 1/2m * сума (ho(х(у степені i)) - y(у степені i)) у квадраті
(сума від і = 1 до m)
де:
J(θ) - функція втрат;
θ - вектор параметрів моделі (вектор ваг);
m - кількість прикладів у навчальному наборі;
ho(х(у степені i)) - прогнозоване значення для i-го прикладу;
y( у степені i) - справжнє значення для i-го прикладу.
Ось функція для обчислення функції втрат у векторному вигляді в Python:
Ця функція обчислює функцію втрат для лінійної регресії за допомогою середньоквадратичної помилки.
 Вона приймає вектор параметрів θ, матрицю ознак X і вектор цільових значень y,
  а повертає значення функції втрат J(θ)
"""
def compute_cost(w, X, y):
    """
    Parameters:
    w (numpy.ndarray): Vector of parameters.
    X (numpy.ndarray): Matrix of features.
    y (numpy.ndarray): Vector of target values.

    Returns:
    float: Cost of the linear regression model.
    """
    m = len(y)  # Number of training examples
    h = linear_hypothesis(w, X)  # Compute predictions
    squared_error = np.square(h - y)  # Compute squared error
    cost = (1 / (2 * m)) * np.sum(squared_error)  # Compute cost
    return cost
```

```python
# реалізуйте один крок градієнтного спуску;
"""
Градієнтний спуск - це ітеративний алгоритм оптимізації,
який шукає мінімум (або максимум) функції, змінюючи параметри по напрямку,
протилежному градієнту функції в точці.
У випадку лінійної регресії, ми шукаємо мінімум функції втрат за допомогою градієнтного спуску.
Один крок градієнтного спуску для лінійної регресії виглядає наступним чином:
θj:= θj - α * 1/m * сума (ho(х(у степені i)) - y(у степені i)) * xj(у степені i)
(сума від і = 1 до m)
де:
θj - j-тий параметр моделі;
α - крок навчання (швидкість навчання);
m - кількість прикладів у навчальному наборі;
ho(х(у степені i)) - прогнозоване значення для i-го прикладу;
y(у степені i) - справжнє значення для i-го прикладу;
xj(у степені i) - j-та ознака для i-го прикладу.
Ось функція для виконання одного кроку градієнтного спуску в Python:
Ця функція приймає поточні параметри моделі θ, матрицю ознак X, вектор цільових значень y
та крок навчання α. Вона обчислює градієнт функції втрат, використовуючи поточні параметри
та потім оновлює параметри моделі θ за допомогою градієнтного спуску.
Нові значення параметрів повертаються як результат.
"""
def gradient_descent(X, y, learning_rate, num_iterations):
    """
    Perform gradient descent to optimize parameters w.

    Parameters:
    X (numpy.ndarray): Matrix of features.
    y (numpy.ndarray): Vector of target values.
    learning_rate (float): Learning rate (step size).
    num_iterations (int): Number of iterations.

    Returns:
    numpy.ndarray: Optimized parameters w.
    """
    m, n = X.shape  # Number of training examples, number of features
    w = np.zeros(n)  # Initialize parameters to zeros

    for _ in range(num_iterations):
        h = linear_hypothesis(w, X)  # Compute predictions
        gradient = (1 / m) * np.dot(X.T, (h - y))  # Compute gradient
        w -= learning_rate * gradient  # Update parameters

    return w
```

```python
# знайдіть найкращі параметри w для датасету використовуючи написані вами функції,
# прогнозуючу ціну на будинок залежно від площі, кількості ванних кімнат та кількості спалень;
df = pd.read_csv("Housing.csv")
df
```

```python
"""
To find the best parameters w for the dataset using linear regression, we can follow these steps:
1 - Prepare the Data: Load the dataset and preprocess it as necessary.
2 - Define Features and Target: Separate the features (independent variables) from the target variable (dependent variable).
3 - Normalize Features: It's often beneficial to normalize the features to ensure that they have similar scales.
4 - Add Bias Term: Add a bias term (intercept) to the features matrix.
5 - Initialize Parameters: Initialize the parameters w.
6 - Define the Cost Function: Use the Mean Squared Error (MSE) as the cost function.
7 - Gradient Descent: Implement gradient descent to minimize the cost function
and find the optimal parameters w.
8 - Train the Model: Run gradient descent iteratively until convergence to find the optimal parameters w.
9 - Evaluate the Model: Evaluate the model on the training data and check the performance.
Let's implement these steps:
This code will find the optimal parameters w using gradient descent for the given dataset,
 predicting house prices based on square footage, number of bathrooms, and number of bedrooms.
 Adjust the learning rate and the number of iterations as needed for better convergence.
"""
# Prepare the data
X = df[['area', 'bathrooms', 'bedrooms']].values
y = df['price'].values

# Normalize features (optional but often beneficial)
X = (X - X.mean(axis=0)) / X.std(axis=0)

# Add bias term (intercept) to features matrix
X = np.column_stack((np.ones(len(X)), X))

# Define hyperparameters
learning_rate = 0.01
num_iterations = 1000

# Run gradient descent to find the optimal parameters
w = gradient_descent(X, y, learning_rate, num_iterations)

print("Optimal parameters (w):", w)
```

```python
# знайдіть ці ж параметри за допомогою аналітичного рішення;
"""
To find the optimal parameters w using analytical solution (also known as the normal equation)
for linear regression, you can use the following formula:
w = (X(у степені Т) * X)(у степені -1) * X(у степені Т) * y
where:
X - the design matrix (including the bias term);
y - the vector of target values.
Here's how you can implement this in Python:
This code will calculate the optimal parameters w using the normal equation for the given dataset.
The resulting parameters w will be printed out.
"""
# Add bias term (intercept) to features matrix
X = np.column_stack((np.ones(len(df)), df[['area', 'bathrooms', 'bedrooms']].values))

# Target vector
y = df['price'].values

# Calculate optimal parameters using the normal equation
w = np.linalg.inv(X.T @ X) @ X.T @ y

print("Optimal parameters (w) using analytical solution:", w)
```

```python
# для перевірки спрогнозованих значень, використайте LinearRegression з бібліотеки scikit-learn та порівняйте результати.
"""
Sure, we can use the LinearRegression model from the scikit-learn library
to make predictions and compare them with the results obtained using the analytical solution.
Here's how you can do it:
This code fits a LinearRegression model to the data
and then retrieves the coefficients (parameters)
and the intercept from the trained model.
We concatenate the intercept with the coefficients to obtain the complete parameter vector w.
Finally, we print out the optimal parameters obtained using scikit-learn's LinearRegression model.
"""
# Create a LinearRegression model
model = LinearRegression()

# Fit the model to the data
model.fit(X[:, 1:], y)  # Exclude the bias term from X

# Get the coefficients (parameters) and intercept from the model
w_sklearn = np.concatenate(([model.intercept_], model.coef_))

print("Optimal parameters (w) using scikit-learn LinearRegression:", w_sklearn)
"""
Now, you can compare the results obtained from the analytical solution
with those obtained using scikit-learn's LinearRegression model
to verify the correctness of the calculations.
"""
```
